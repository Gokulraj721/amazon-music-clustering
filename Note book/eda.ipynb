```python
selected_features = [
    'popularity_songs', 'duration_ms', 'danceability', 'energy',
    'key', 'loudness', 'mode', 'speechiness', 'acousticness',
    'instrumentalness', 'liveness', 'valence', 'tempo',
    'time_signature', 'followers', 'popularity_artists'
]

df_selected = df[selected_features]
df_selected.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_selected)

scaled_df = pd.DataFrame(scaled_data, columns=selected_features)
scaled_df.head()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia = []
K = range(1, 11)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_df)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(K, inertia, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(scaled_df)

# Assign cluster labels
scaled_df['cluster'] = kmeans.labels_
scaled_df.head()
```

```python
scaled_df['cluster'].value_counts()
```

```python
df['cluster'] = kmeans.labels_
df.head()
```

```python
from sklearn.decomposition import PCA
import seaborn as sns
import matplotlib.pyplot as plt

pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_df.drop('cluster', axis=1))

plt.figure(figsize=(8, 6))
sns.scatterplot(x=pca_data[:, 0], y=pca_data[:, 1], hue=scaled_df['cluster'], palette='Set2')
plt.title('K-Means Clusters (PCA-reduced)')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

cluster_summary = df.groupby('cluster').mean()

cluster_summary.T.plot(kind='bar', figsize=(14, 6))
plt.title('Average Feature Values per Cluster')
plt.ylabel('Scaled Value')
plt.xlabel('Audio Features')
plt.legend(title='Cluster')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.heatmap(cluster_summary, annot=True, cmap='coolwarm')
plt.title('Cluster Feature Heatmap')
plt.xlabel('Features')
plt.ylabel('Cluster')
plt.tight_layout()
plt.show()
```

```python
features_to_plot = ['danceability', 'energy', 'valence']

for feature in features_to_plot:
    plt.figure(figsize=(6, 4))
    sns.histplot(data=df, x=feature, hue='cluster', kde=True, palette='Set2')
    plt.title(f'{feature.capitalize()} Distribution by Cluster')
    plt.tight_layout()
    plt.show()
```
